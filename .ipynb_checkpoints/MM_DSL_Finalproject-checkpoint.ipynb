# LOADING DATASETS

```python
import os
folder = os.getcwd()
slash='\\'
folderData = 'datasets'
fileDev = 'development.csv'
fileEval = 'evaluation.csv'

import it_core_news_sm 
import csv
import numpy as np
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.snowball import SnowballStemmer
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords as sw
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import normalize

def clean_punctuation(text):
    for p in (string.punctuation + "'" +"\n"):
        text = text.replace(p,' ')
    return text
    
dev_reviews = []
with open(folder + slash + folderData + slash + fileDev,'r', encoding="utf-8") as f:
    for row in csv.reader(f):
        row[0] = clean_punctuation(row[0])
        dev_reviews.append(row)
    dev_reviews.pop(0)

eval_reviews = []
with open(folder + slash + folderData + slash + fileEval,'r', encoding="utf-8") as f:
    for row in csv.reader(f):
        row[0] = clean_punctuation(row[0])
        eval_reviews.append(row[0])
    eval_reviews.pop(0)
    
labels = [row[1] for row in dev_reviews]
dev_reviews = [row[0] for row in dev_reviews]

```

# DATA EXPLORATION

```python
label_counts = np.unique(labels, return_counts = True)

def listToString(lista):
    s = ''
    for word in lista:
        s += word + ' '
    
    return s
    

pos_lengths = []
neg_lengths = []
pos_reviews = []
neg_reviews = []
pos_labels = []
neg_labels = []
for r,l in zip(dev_reviews, labels):
    if(l == 'pos'):
        pos_lengths.append(len(r.split()))
        pos_reviews.append(r)
        pos_labels.append(l)
    if(l == 'neg'):
        neg_lengths.append(len(r.split()))
        neg_reviews.append(r)
        neg_labels.append(l)
            
pos_lengths = np.array(pos_lengths)
neg_lengths = np.array(neg_lengths)

print("Positive mean: ", np.mean(pos_lengths))
print("Negative mean: ", np.mean(neg_lengths))
```

# TOKENIZATION - STEMMING - LEMMATIZATION

```python
from nltk.stem.snowball import SnowballStemmer

it_stop_words = stopwords.words('italian')
lemmatizer = it_core_news_sm.load()
stemmer = SnowballStemmer('italian')
def listToString(lista):
    s = ''
    for word in lista:
        s += word + ' '
    
    return s

def extract_words(text):
    clean_text=[]
    for w in text:
        if(( len(w)<=2)  or (bool(re.search(r'\d', w)) == True) ):
            continue
        clean_text.append(w.lower())
    return clean_text

def clean_lemmas(text,stopwords):
    lemmatized_text=[]
    other_stopwords=stopwords + ['altro','stare','trovare','soggiornare','poco','volere','fare','stare','dire','colazione','potere','solere','camera',
                    'hotel', 'albergo','stanza','struttura','per','su','tra','fra']
    for w in text:
        if ((w.pos_ not in {'PUNCT','DET','ADP'}) and ( w.lemma_ not in other_stopwords)):
            lemmatized_text.append(w.lemma_)  
    return listToString(lemmatized_text);

lemmatized_reviews=[]
for review in dev_reviews + eval_reviews:
    word_tokenized=word_tokenize(review)
    word_tokenized_extracted =extract_words(word_tokenized)
    text = stemmer.stem(listToString(word_tokenized_extracted))
    lemmatized_doc = lemmatizer(text)
    lemmatized_reviews.append(clean_lemmas(lemmatized_doc,it_stop_words))
print(lemmatized_reviews[0])
```

# TEXT TOKENIZATION - STEMMING

```python
from nltk.stem.snowball import SnowballStemmer

it_stop_words = stopwords.words('italian')
stemmer = SnowballStemmer('italian')
def listToString(lista):
    s = ''
    for word in lista:
        s += word + ' '
    
    return s

def extract_words(text):
    clean_text=[]
    for w in text:
        if(( len(w)<=2)) or (bool(re.search(r'\d', w)) == True) :
            continue
        clean_text.append(w.lower())
    return clean_text

def clean_lemmas(text,stopwords):
    lemmatized_text=[]
    other_stopwords=stopwords + ['stare','volere','fare','stare','dire','per','su','tra','fra','solo','dopo','dato',
                                'napoli','milano','venezia','perchè','hotel','albergo','camera','struttura','colazione',
                                 'stato','fatto','stanza','quindi','poi','così','quando','stati','stata']
    for w in text.split(' '):
        if ( w not in other_stopwords):
            lemmatized_text.append(w)  
    return listToString(lemmatized_text);

def stem_and_clean(tokens,stopwords):
    stemmed_text=[]
    stopwords=stopwords + ['stare','volere','fare','stare','dire','per','su','tra','fra','solo','dopo','dato',
                                'napoli','milano','venezia','perchè','hotel','albergo','camera','struttura','colazione',
                                 'stato','fatto','stanza','quindi','poi','così','quando','stati','stata']
    for w in tokens:
        if w not in stopwords:
            w = stemmer.stem(w)
            if w not in {'prim','cam','trov','arriv'}:
                stemmed_text.append(w)
    return listToString(stemmed_text)

lemmatized_reviews=[]
for review in dev_reviews + eval_reviews:
    word_tokenized=word_tokenize(review)
    word_tokenized_extracted =extract_words(word_tokenized)
    lemmatized_reviews.append(stem_and_clean(word_tokenized_extracted,it_stop_words))
print(lemmatized_reviews[0])
```

# TOKENIZATION - STEMMING - LEMMATIZATION - UNDERSAMPLING

```python
from nltk.stem.snowball import SnowballStemmer

it_stop_words = stopwords.words('italian')
lemmatizer = it_core_news_sm.load()
stemmer = SnowballStemmer('italian')
def listToString(lista):
    s = ''
    for word in lista:
        s += word + ' '
    
    return s

def extract_words(text):
    clean_text=[]
    for w in text:
        if(( len(w)<=2)  or (bool(re.search(r'\d', w)) == True) ):
            continue
        clean_text.append(w.lower())
    return clean_text
        
def remove_stopwords(text,stopwords):
    clean_text=[]
    for w in text:
        if w not in stopwords:
            clean_text.append(w)
    
    return clean_text

def clean_lemmas(text):
    lemmatized_text=[]
    other_stopwords={'altro','stare','trovare','soggiornare','poco','volere','fare','stare','dire','potere','solere','camera',
                    'hotel', 'albergo','stanza','struttura'}
    for w in text:
        if ((w.pos_ not in {'PUNCT','DET','ADP'}) and ( w.lemma_ not in other_stopwords)):
            lemmatized_text.append(w.lemma_)  
    return listToString(lemmatized_text);

lemmatized_reviews_US=[]
for review in undersampling_dev_reviews + eval_reviews:
    word_tokenized=word_tokenize(review)
    word_tokenized_extracted =extract_words(word_tokenized)
    word_tokenized_no_sw =remove_stopwords(word_tokenized_extracted,it_stop_words)
    text = listToString(word_tokenized_no_sw)
    text = stemmer.stem(text)
    lemmatized_doc = lemmatizer(text)
    lemmatized_reviews_US.append(clean_lemmas(lemmatized_doc))
print(lemmatized_reviews_US[0])
```

# TRAINING - LINEAR SVC

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
def gridSearchCV_SVC(Reviews_train,reviews_labels):
    classifier = LinearSVC(fit_intercept=True,random_state=42,tol=1e-5)
    param_grid = { 
                  'C':[1.5,2.0,2.5],
                  'penalty' : ['l2'],
                  'loss' : ['hinge','squared_hinge'],
                  'class_weight' : ['balanced',None]
                  
                   
                   }
    
    gridsearch = GridSearchCV(classifier, param_grid, scoring='f1_weighted', cv=5)
    gridsearch.fit(Reviews_train,reviews_labels)
    return gridsearch
    
Reviews_train,Reviews_test,reviews_labels,test_labels= train_test_split(tfidf_reviews_dev,labels,train_size=0.85,stratify=labels,shuffle=True)
gridsearchCV_SVC_=gridSearchCV_SVC(Reviews_train,reviews_labels)

y_pred=gridsearchCV_SVC_.best_estimator_.predict(Reviews_test)
print(gridsearchCV_SVC_.best_estimator_)
print(gridsearchCV_SVC_.best_score_)
print(f1_score(y_pred,test_labels,average = 'weighted'))
print(classification_report(y_pred,test_labels))
```
