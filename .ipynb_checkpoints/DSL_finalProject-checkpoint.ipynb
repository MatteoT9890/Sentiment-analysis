# Data loading and first cleaning

```python
import os
folder = os.getcwd()
slash='\\'
folderData = 'datasets'
fileDev = 'development.csv'
fileEval = 'evaluation.csv'

import csv
import numpy as np
from pattern.text import it as pattern_it 
import string
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from nltk.corpus import stopwords as sw
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
from sklearn.preprocessing import normalize

def clean_punctuation(text):
    for p in (string.punctuation + "'" +"\n"):
        text = text.replace(p,' ')
    return text
    
dev_reviews = []
with open(folder + slash + folderData + slash + fileDev,'r', encoding="utf-8") as f:
    for row in csv.reader(f):
        row[0] = clean_punctuation(row[0])
        dev_reviews.append(row)
    dev_reviews.pop(0)

eval_reviews = []
with open(folder + slash + folderData + slash + fileEval,'r', encoding="utf-8") as f:
    for row in csv.reader(f):
        row[0] = clean_punctuation(row[0])
        eval_reviews.append(row[0])
    eval_reviews.pop(0)
    
labels = [row[1] for row in dev_reviews]
dev_reviews = [row[0] for row in dev_reviews]

```

# Text tokenization and lemmatization

```python
it_stop_words = stopwords.words('italian') + ['hotel', 'albergo', 'camera','stanza', 'struttura','personale','staff']

def lemmatize_word(word):
    in_word = word#.decode('utf-8')
    # print('Something: {}'.format(in_word))

    word_it = pattern_it.parse(
        in_word, 
        tokenize=False,  
        tag=False,  
        chunk=False,  
        lemmata=True 
    )
    # print("Input: {} Output: {}".format(in_word, word_it))
    the_lemmatized_word = word_it.split()[0][0][4]
    # print("Returning: {}".format(the_lemmatized_word))
    return the_lemmatized_word

def extract_words(text):
    clean_text=[]
    for w in text:
        if(( len(w)<=2)  or (bool(re.search(r'\d', w)) == True) ):
            continue
        clean_text.append(w)
    return clean_text
        
def remove_stopwords(text,stopwords):
    clean_text=[]
    for w in text:
        if w not in stopwords:
            clean_text.append(w)
    
    return clean_text

def lemmatize(text):
    lemmatized_text=[]
    for w in text:
        lemmatized_text.append(lemmatize_word(w))
    
    return lemmatized_text;

lemmatized_reviews=[]
for review in dev_reviews + eval_reviews:
    word_tokenized=word_tokenize(review)
    word_tokenized_extracted =extract_words(word_tokenized)
    word_tokenized_no_sw =remove_stopwords(word_tokenized_extracted,it_stop_words)
    lemmatized_text = lemmatize(word_tokenized_no_sw)
    lemmatized_reviews.append(lemmatized_text)


print(lemmatized_reviews[0])
```

# Algorithm selection

```python
%matplotlib notebook

lemmatized_reviews_string = [listToString(review) for review in lemmatized_reviews]
reviews_vectorizer = TfidfVectorizer(min_df=5,max_df=0.5)
tfidf_reviews = reviews_vectorizer.fit_transform(lemmatized_reviews_string)

tfidf_reviews_dev = tfidf_reviews[:len(dev_reviews),:]
tfidf_reviews_eval=tfidf_reviews[len(dev_reviews):,:]

svd = TruncatedSVD(n_components=10,random_state=41)
svd = svd.fit(tfidf_reviews)
tfidfSVD = svd.transform(tfidf_reviews_dev)
plt.plot(svd.explained_variance_ratio_, linestyle='-')
```

```python
from mpl_toolkits.mplot3d import Axes3D
svd_plot = TruncatedSVD(n_components=3,random_state=41)
svd_plot = svd_plot.fit(tfidf_reviews)
tfidfSVD_plot = svd_plot.transform(tfidf_reviews_dev)
tfidfSVD_plot = normalize(tfidfSVD_plot)
fig=plt.figure()
ax = Axes3D(fig)

num_labels = []
for l in labels:
    if(l == 'pos'):
        num_labels.append(1)
    else:
        num_labels.append(0)
        
# Data for a three-dimensional line
ax.scatter(tfidfSVD_plot[:,0],tfidfSVD_plot[:,1] ,tfidfSVD_plot[:,2],c=num_labels,s=0.1)
```

# TRAINING - TFIDF

```python
def gridSearchCV(Reviews_train,reviews_labels):
    classifier = RandomForestClassifier(random_state = 41,max_features='sqrt')
    param_grid = {  
                   'n_estimators' : [100]}
    
    gridsearch = GridSearchCV(classifier, param_grid, scoring='f1_weighted', cv=5)
    gridsearch.fit(Reviews_train,reviews_labels)
    return gridsearch
    
Reviews_train,Reviews_test,reviews_labels,test_labels= train_test_split(tfidf_reviews_dev,labels,train_size=0.75)

gridsearchCV=gridSearchCV(Reviews_train,reviews_labels)
y_pred=gridsearchCV.best_estimator_.predict(Reviews_test)
print(gridsearchCV.best_estimator_)
print(gridsearchCV.best_score_)
#classifier = RandomForestClassifier(n_estimators = 60,random_state = 41,max_features=None)
#y_pred = classifier.fit(Reviews_train,reviews_labels).predict(Reviews_test)
f1_score(y_pred,test_labels,average = 'weighted')
```
