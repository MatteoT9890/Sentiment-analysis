# LOADING DATASET

```python
import os
folder = os.getcwd()
slash='\\'
folderData = 'datasets'
fileDev = 'development.csv'
fileEval = 'evaluation.csv'

import csv
import numpy as np

reviews = []
with open(folder + slash + folderData + slash + fileDev,'r', encoding="utf-8") as f:
    for row in csv.reader(f):
        reviews.append(row)
    reviews.pop(0)
```

# DATA EXPLORATION

```python
labels = [row[1] for row in reviews]
reviews = [row[0] for row in reviews]

np.unique(labels, return_counts = True)
```

# DATA PREPROCESSING

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
it_stop_words = stopwords.words('italian')
from pattern.text import it as pattern_it 
import string
import re
def lemmatize_word(word):
    in_word = word#.decode('utf-8')
    # print('Something: {}'.format(in_word))
    
    word_it = pattern_it.parse(
        in_word, 
        tokenize=False,  
        tag=False,  
        chunk=False,  
        lemmata=True 
    )    
    # print("Input: {} Output: {}".format(in_word, word_it))
    the_lemmatized_word = word_it.split()[0][0][4]
    # print("Returning: {}".format(the_lemmatized_word))
    return the_lemmatized_word

def clean_punctuation(text):
    clean_text=[]
    for w in text:
        if(( len(w)<=2)  or (bool(re.search(r'\d', w)) == True) or (w in string.punctuation)):
            continue
        clean_text.append(w.lower().split("'")[0])
    return clean_text
        
def remove_stopwords(text,stopwords):
    clean_text=[]
    for w in text:
        if w not in stopwords:
            clean_text.append(w)
    
    return clean_text

def lemmatize(text):
    lemmatized_text=[]
    for w in text:
        lemmatized_text.append(lemmatize_word(w))
    
    return lemmatized_text;

lemmatized_reviews=[]
for review in reviews:
    word_tokenized=word_tokenize(review)
    word_tokenized_no_punct = clean_punctuation(word_tokenized)
    word_tokenized_no_punct_no_sw =remove_stopwords(word_tokenized_no_punct,it_stop_words)
    lemmatized_text = lemmatize(word_tokenized_no_punct_no_sw)
    lemmatized_reviews.append(lemmatized_text)
print(lemmatized_reviews[0])
```
